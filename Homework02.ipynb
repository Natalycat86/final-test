```python
import pandas as pd
import numpy as np


#impor csv file
df = pd.read_csv('./ab_stats.csv')

df.head()
#df.info()

#arppu = revenue / num_purchases
# arppu for a group
arppu_a = df[df['ab_group'] == 'A']['revenue'].sum() / df[df['ab_group'] == 'A']['num_purchases'].sum()
arppu_b = df[df['ab_group'] == 'B']['revenue'].sum() / df[df['ab_group'] == 'B']['num_purchases'].sum()
print(arppu_a)
print(arppu_b)
print(f"Разница ARPPU: {arppu_a - arppu_b}")

#стат значимость ab_test
from scipy import stats
stats.ttest_ind(df[df['ab_group'] == 'A']['revenue'], df[df['ab_group'] == 'B']['revenue'], equal_var=False)
print(stats.ttest_ind(df[df['ab_group'] == 'A']['revenue'], df[df['ab_group'] == 'B']['revenue'], equal_var=False))



# Рекомендации менеджерам: arppu_a > arppu_b на 1.26, При статистической значимости 1.26.
```

5.  Мы провели АБ-тест на увеличение average timespent per user. По итогам тестирования мы 
получили следующие данные. Является ли результат статистически значимым с уровнем 
доверия 80%? Какую версию мы выкатим на продакшн?
A) Средняя - 360, отклонение - 40, количество - 9802
B) Средняя - 352, отклонение - 58, количество - 9789

### Ответ


Чтобы определить, является ли результат статистически значимым, нам нужно рассчитать доверительный интервал для разницы между средними значениями двух групп. Поскольку уровень доверия составляет 80%, нам нужно рассчитать доверительный интервал с уровнем значимости 20% (1 - 0,8 = 0,2).

**Расчет доверительного интервала**

Для расчета доверительного интервала мы можем использовать формулу:

Доверительный интервал = (среднее значение A - среднее значение B) ± (Z-коэффициент \* стандартная ошибка)

где Z-коэффициент соответствует уровню значимости 20% (Z = 1,28 для двухстороннего теста).

**Расчет стандартной ошибки**

Стандартная ошибка (SE) рассчитывается как:

SE = sqrt((отклонение A^2 / количество A) + (отклонение B^2 / количество B))

SE = sqrt((40^2 / 9802) + (58^2 / 9789)) ≈ 2,33

**Расчет доверительного интервала**

Доверительный интервал = (360 - 352) ± (1,28 \* 2,33) ≈ 8 ± 2,98

Доверительный интервал ≈ (5,02; 10,98)

**Вывод**
Поскольку доверительный интервал не пересекает ноль, мы можем заключить, что разница между средними значениями двух групп статистически значима с уровнем доверия 80%.

**Рекомендация**
Выкатить версию A на продакшн, поскольку она показала более высокое среднее значение времени, проведенного пользователем (360 против 352).


```python
import math

# Define the data
mean_A = 360
stddev_A = 40
n_A = 9802

mean_B = 352
stddev_B = 58
n_B = 9789

# Calculate the standard error
SE = math.sqrt((stddev_A**2 / n_A) + (stddev_B**2 / n_B))

# Calculate the Z-score
Z_score = (mean_A - mean_B) / SE

# Calculate the confidence interval
confidence_level = 0.8
alpha = 1 - confidence_level
Z_critical = 1.28  # for a two-tailed test with alpha = 0.2
CI = (mean_A - mean_B) - Z_critical * SE, (mean_A - mean_B) + Z_critical * SE

print("Confidence Interval:", CI)

# Check if the result is statistically significant
if CI[0] > 0:
    print("Результат является статистически значимым с уровнем достоверности 80%.")
    print("Мы рекомендуем запустить версию А в производство.")
else:
    print("Результат не является статистически значимым с уровнем достоверности 80%.")
    print("Мы не можем сделать вывод, что версия A лучше, чем версия B.")

```
